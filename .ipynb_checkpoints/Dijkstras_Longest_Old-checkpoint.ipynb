```python
#Second
class Dijkstras:
    def __init__(self, data, matrix):
        self.data = data
        self.matrix = matrix
        
    def longest(self, source, target0, target1):
        n = len(self.matrix[source])
        
        #set E2E Transimition rate as E2E_T in -inf
        E2E_T = np.full(n, (-1*np.inf))  
        E2E_T = np.array([E2E_T],).tolist()
        E2E_T[0][source] = 0
        
        prev_node = np.full(n, -1)
        prev_node = np.array([prev_node],).tolist()

        visited = np.full(n, 0)
        visited = np.array([visited],).tolist()
        visited[0][source] = 1
    
        more = True
        
        fork = 0
        
        node_i = source
        
        print_route = []
    
                
        while (more):

            
            max_g = (-1*np.inf)
            
            for idx, next_node in enumerate(E2E_T[fork]):
                if visited[fork][idx] != 1:
                    if self.matrix[node_i][idx] != 0 and self.matrix[node_i][idx] > next_node:
                        E2E_T[fork][idx] = self.matrix[node_i][idx]
                        prev_node[fork][idx] = node_i
                        
                    if self.matrix[node_i][idx] > max_g:
                        max_g = self.matrix[node_i][idx]
        
            
            fork_i = 0
            for idx, next_node in enumerate(E2E_T[fork]):
                if visited[fork][idx] != 1:
                    if next_node == max_g:
                        fork_i = fork_i + 1
            
            E2E_T = E2E_T * fork_i
            prev_node = prev_node * fork_i
            visited = visited * fork_i
            
            for x in range(fork_i):
                temp = []
                for idx, next_node in enumerate(E2E_T[x]):
                    if visited[x][idx] != 1 and next_node not in temp :
                        if next_node == max_g:
                            visited[fork][idx] = 1
                            temp.append(next_node)
                            print_route.append([self.data[idx],E2E_T[fork][idx]])
                            node_i = idx
                            
                        if (node_i == target0 or node_i == target1):
                            more = False
            
        min_E2E = min(np.array(print_route)[:,1])

        p1 = {"source": self.data[source], "routing path": print_route, "End-to-end rate": min_E2E}

        return p1

        
aa = Dijkstras(data = data, matrix = tr_matrix)
aa.longest(source = 4, target0 = 6, target1 = 7)   
```
